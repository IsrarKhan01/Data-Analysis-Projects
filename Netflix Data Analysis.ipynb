### Q.1 In which year highest number or TV Shows and Movies were released? Show with the help of Bargraph? 

```python
df.dtypes
```

```python
## the column that we require for this task is in Object form so first we have to convert it into date time format
```

```python
df["Release_Dt"]=pd.to_datetime(df["Release_Date"])
```

```python
df.dtypes
```

```python
df["Release_Dt"].dt.year.value_counts()
```

#### As we can see that the YEAR which has the highest TV and Movies Releases is 2019 followed by 2020, 2018, 2017.

```python
import seaborn as sns
df["Release_Dt"].dt.year.value_counts().plot(kind="bar")
```
